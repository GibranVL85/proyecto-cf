```python
#Importacion de librerias
import altair as alt
import pandas as pd 
```

```python
#Carga de datasource
df = pd.read_csv('../Datasource/ufcFigths_010824_ver2.csv')

df['date'] = pd.to_datetime(df['date'])

#Validacion de la informacion del CSV
df
```

## Revisi√≥n de la data y generacion de sampleo

```python

#Sampleo de data con asignacion aleatoria
sampled_data = df.sample(n=5000, random_state=1) 

# visualizacion de Chart con la prueba de datos
print(sampled_data)
```

# Analisis sobre el m√©todo de decisi√≥n
Se realiza una revision sobre el historico de las peleas para determinar cual es el metodo de decision mas comun.

## Data para Gr√°ficos


```python

#Data donde se escogen los valores de a√±o [x] y el m√©todo de decisi√≥n[y]
chartData = pd.DataFrame({
    'x': sampled_data['date'].dt.year,
    'y': sampled_data.decision_method
})
print(chartData)

# Nueva data donde se agrupan los valores distintos de cada metodo de decision por a√±o
count_decision_met = chartData.groupby(['y']).size().reset_index(name='count')
print("Total de metodos de decision")
print(count_decision_met)


# Nueva data donde se agrupan los valores distintos de cada metodo de decision por a√±o
grouped_count = chartData.groupby(['x', 'y']).size().reset_index(name='count') 
print("Ajuste en la data para visualizar los metodos de decision por a√±o")
print(grouped_count)


```

## Gr√°ficos de m√©todo de decisi√≥n

```python
#Gr√°fica de barras en altair
bars = alt.Chart(grouped_count).mark_bar().encode(
    y=alt.Y('y', title='M√©todos de Decisi√≥n'),   
    x=alt.X('count:Q', title='Conteo de registros'), 
    
    color='y:N', # se agrega color para identificar cada metodo de decision
   
).properties(
    title='Total de Conteo de m√©todos de decision'
).interactive()


#Mostrar gr√°fico
bars
```

```python

#Gr√°fica de barras en altair
bar_chart = alt.Chart(grouped_count).mark_bar().encode(
    x=alt.X('x:O', title='A√±o'),   #A√±o como temporal
    y=alt.Y('count:Q', title='M√©todos de Decisi√≥n'),   #Conteo como tipo cuantitativo
    color='y:N', # se agrega color para identificar cada metodo de decision
    tooltip=['x', 'y', 'count']  
).properties(
    title='Conteo de m√©todos de decision por A√±o'
).interactive()


#Mostrar gr√°fico
bar_chart
```

# Analisis Exploratorio de Datos con visualizaci√≥n efectiva
Se realiza una revision sobre los golpes recibidos, el objetivo es identificar a los peleadores mas golpeados

## Data para Gr√°ficos

```python
#Seleccion de columnas para su analisis

df = pd.DataFrame({
    'date':pd.to_datetime(sampled_data.date), #Fecha de combate
    'decision_method':sampled_data.decision_method, #metodo de decisi√≥n
    'winner':sampled_data.winner, #Nombre del peleador ganador del combate
    'fight_id':sampled_data.fight_id, #ID del combate (pelea)
    'fighter_1':sampled_data.fighter_1, #Nombre del peleador 1
    'fighter_2':sampled_data.fighter_2, #Nombre del peleador 2
    'body_abs_nb':sampled_data['body_abs'].str[1:-1], # Golpes recibidos en el cuerpo por el Peleador 1 y el Peleador 2 Se quitan los parentesis cuadrados a la data
    'head_abs_nb':sampled_data['head_abs'].str[1:-1], # Golpes recibidos en la cabeza por el Peleador 1 y el Peleador 2
    'leg_abs_nb':sampled_data['leg_abs'].str[1:-1], # Golpes recibidos en las piernas por el Peleador 1 y el Peleador 2
    'distance_abs_nb':sampled_data['distance_abs'].str[1:-1], # Golpes recibidos a la distancia por el Peleador 1 y el Peleador 2
    'sig_str_abs_nb':sampled_data['sig_str_abs'].str[1:-1] # Total de golpes significativos recibidos por el Peleador 1 y el Peleador 2
    
})


```

## Limpieza de columnas
Proceso de limpiado de datos ya que estos vienen en un formato [x:y], donde:
     x = al numero de golpes que recibio en la pelea el Peleador 1
     y = al numero de golpes que recibio en la pelea el Peleador 2
Este formato lo utilizan todas las columnas que reportan golpes, de igual forma se pueden identificar con la terminacion '_abs'

```python


df[['body_abs_1', 'body_abs_2']] = df['body_abs_nb'].str.split(', ', expand=True)
df[['head_abs_1', 'head_abs_2']] = df['head_abs_nb'].str.split(', ', expand=True)
df[['leg_abs_1', 'leg_abs_2']] = df['leg_abs_nb'].str.split(', ', expand=True)
df[['distance_abs_1', 'distance_abs_2']] = df['distance_abs_nb'].str.split(', ', expand=True)
df[['sig_str_abs_1', 'sig_str_abs_2']] = df['sig_str_abs_nb'].str.split(', ', expand=True)

```

## Orden de los dataframes
Se tuvo que crear 2 Dataframes para despues unirlos en 1 solo, esto debido a que 1 combate cuanta con la informacion de 2 peleadores el peleador 1 y el peleador 2 y cada uno recibio golpes durante la pelea. Es decir la informacion fuente se encontraba de la siguiente forma:

date, decision_method, winner, fighter_1, fighter_2, body_abs, head_abs, leg_abs, distance_abs, sig_str_abs

Siendo las columnas *_abs las que contaban con informacion doble del peleador 1 y el peleador 2 de la siguiente forma [x:y]. Esto complicaba el analisis de los datos por lo cual se procedio a separa la informaci√≥n en 2 dataframes.



```python
df_figther1 = df[['fight_id','date','fighter_1','body_abs_1','head_abs_1','leg_abs_1','distance_abs_1','sig_str_abs_1']]
df_figther2 = df[['fight_id','date','fighter_2','body_abs_2','head_abs_2','leg_abs_2','distance_abs_2','sig_str_abs_2']]
df_figther2 = df_figther2.rename(columns={"fighter_2": "fighter_1", "body_abs_2": "body_abs_1",
                                         "head_abs_2": "head_abs_1", "leg_abs_2": "leg_abs_1",
                                         "distance_abs_2": "distance_abs_1", "sig_str_abs_2": "sig_str_abs_1",})

df_fighters = pd.concat([df_figther1,df_figther2])
df_fighters = df_fighters.reset_index(drop=True)

#print(df_fighters.sort_values('head_abs_1',ascending=False))
#print(df_figther2)
print(df_fighters)
```

## Dataframes agrupados
Con los dataframes agrupados es factible realizar el analisis de la informacion y poder identificar los siguientes puntos:
    * Top 10 peleadores con mas peleas en la UFC
    * Top 20 peleadores con m√°s golpes recibidos en el cuerpo
    * Top 20 peleadores con m√°s golpes recibidos en la cabeza
    * Top 20 peleadores con m√°s golpes recibidos en las piernas
    * Top 20 peleadores con m√°s golpes significativos recibidos



### Top 10 peleadores con mas peleas en la UFC

```python
count_figths = df_fighters.groupby(['fighter_1']).size().reset_index(name='count_figths')
count_figths = count_figths.sort_values('count_figths',ascending=False)
print("Top 10 peleadores con mas peleas en la UFC")
print(count_figths.head(10))
```

```python
#Gr√°fica

chart = alt.Chart(count_figths.head(10)).mark_bar().encode(
    x=alt.X('count_figths:Q', title='N√∫mero de peleas'),
    y=alt.Y('fighter_1:N', sort='-x', title='Peleador'),
    color=alt.Color('count_figths:Q', scale=alt.Scale(scheme='blues'))  # Colorear las barras
).properties(
    title='Top 10 peleadores con m√°s peleas en la UFC'
)

# Mostrar la gr√°fica
chart
```

### Top 20 peleadores con m√°s golpes recibidos en el cuerpo

```python

df_fighters['body_abs_1'] = pd.to_numeric(df_fighters['body_abs_1'], errors='coerce')
body_hits_sum = df_fighters.groupby('fighter_1')['body_abs_1'].sum().reset_index()
body_hits_sum = body_hits_sum.sort_values('body_abs_1', ascending=False)

print("Top 20 peleadores con m√°s golpes recibidos en el cuerpo")
print(body_hits_sum.head(20))


```

```python
#Gr√°fica
chart = alt.Chart(body_hits_sum.head(25)).mark_bar().encode(
    x=alt.X('fighter_1:N', sort='-y', title='Peleador'),
    y=alt.Y('body_abs_1:Q', title='N√∫mero de peleas'),
    color=alt.Color('body_abs_1:Q', scale=alt.Scale(scheme='greens'))
).properties(
    title='Top 20 peleadores con m√°s golpes recibidos en el cuerpo'
).interactive()

chart
```

### Top 20 peleadores con m√°s golpes recibidos en la cabeza

```python

df_fighters['head_abs_1'] = pd.to_numeric(df_fighters['head_abs_1'], errors='coerce')
head_hits_sum = df_fighters.groupby('fighter_1')['head_abs_1'].sum().reset_index()
head_hits_sum = head_hits_sum.sort_values('head_abs_1', ascending=False)

print("Top 20 peleadores con m√°s golpes recibidos en la cabeza")
print(head_hits_sum.head(20))
```

```python
#Gr√°fica

chart = alt.Chart(head_hits_sum.head(25)).mark_point().encode(
    x=alt.X('head_abs_1:Q', title='N√∫mero de peleas'),
    y=alt.Y('fighter_1:N', sort='-x', title='Peleador'),
    size='head_abs_1:Q',  # Tama√±o del punto seg√∫n n√∫mero de peleas
    color=alt.Color('head_abs_1:Q', scale=alt.Scale(scheme='purples')),
     tooltip=['fighter_1', 'head_abs_1']  
).properties(
    title='Top 20 peleadores con m√°s golpes recibidos en la cabeza'
).interactive()

chart
```

### Top 20 peleadores con m√°s golpes recibidos en las piernas

```python
df_fighters['leg_abs_1'] = pd.to_numeric(df_fighters['leg_abs_1'], errors='coerce')
leg_hits_sum = df_fighters.groupby('fighter_1')['leg_abs_1'].sum().reset_index()
leg_hits_sum = leg_hits_sum.sort_values('leg_abs_1', ascending=False)

print("Top 20 peleadores con m√°s golpes recibidos en las piernas")
print(leg_hits_sum.head(20))
```

```python
#Gr√°fica
chart = alt.Chart(leg_hits_sum.head(20)).mark_arc().encode(
    theta=alt.Theta('leg_abs_1:Q', title='N√∫mero de peleas'),
    color=alt.Color('fighter_1:N', title='Peleador'),
     tooltip=['fighter_1', 'leg_abs_1']  
).properties(
    title='Top 20 peleadores con m√°s golpes recibidos en las piernas'
).interactive()

chart
```

### Top 20 peleadores con m√°s golpes significativos recibidos

```python
df_fighters['sig_str_abs_1'] = pd.to_numeric(df_fighters['sig_str_abs_1'], errors='coerce')
sig_hits_sum = df_fighters.groupby('fighter_1')['sig_str_abs_1'].sum().reset_index()
sig_hits_sum = sig_hits_sum.sort_values('sig_str_abs_1', ascending=False)

print("Top 20 peleadores con m√°s golpes significativos recibidos")
print(sig_hits_sum.head(20))
```

```python
#Gr√°fica
chart = alt.Chart(sig_hits_sum.head(20)).mark_line(point=True).encode(
    x=alt.X('fighter_1:N', sort='-y', title='Peleador'),
    y=alt.Y('sig_str_abs_1:Q', title='N√∫mero de peleas'),
    color=alt.Color('fighter_1:N', title='Peleador'),
    tooltip=['fighter_1', 'sig_str_abs_1']
).properties(
    title='Top 20 peleadores con m√°s golpes significativos recibidos'
).interactive()

chart
```
