```python
#Importacion de librerias
import pandas as pd
from ydata_profiling import ProfileReport
```

## Creación de dataframe

```python
#Carga de datasource
df = pd.read_csv('../Datasource/ufcFigths_010824_ver2.csv')

#Ajustes de datos
df['date'] = pd.to_datetime(df['date'])

#Creacion de Dataframe para trabajar
df_EDA = df
```

```python
#Validacion de la informacion del dataframe a trabajar
df_EDA.info()
```

```python
#Tamaño actual del DataFrame
len(df_EDA)

```

## Preparación y Limpieza de datos

```python
#limpieza de nulls

columnas_limpieza = ['body_abs','body_att','body_def','body_land','clinch_abs','clinch_att',
                     'clinch_def','clinch_land','distance_abs','distance_att','distance_def',
                     'distance_land','ground_abs','ground_att','ground_def','ground_land',
                     'head_abs','head_att','head_def','head_land','kd','leg_abs','leg_att',
                     'leg_def','leg_land','n_pass','n_rev','n_sub','sig_str_abs','sig_str_att',
                     'sig_str_def','sig_str_land','td_abs','td_att','td_def','td_land',
                     'total_str_abs','total_str_att','total_str_def','total_str_land']

df_EDA.dropna(subset=columnas_limpieza, inplace=True)


```

```python
#Validacion de tamaño
len(df_EDA)
```

```python

# Asegurarse de que los valores no contengan basura y se limpia los parentesis en las columnas con formato [x,y]
for col in columnas_limpieza:
    df_EDA[col] = df_EDA[col].apply(lambda x: x[1:-1] if isinstance(x, str) and x.startswith('[') and x.endswith(']') else None)
   
```

```python
#Validacion de la Data sin parentesis cuadrado
df_EDA.head(5)
```

```python
# Crear un DataFrame vacío para almacenar las nuevas columnas
new_columns = pd.DataFrame()

 # Dividir los valores de las columnas, solo si el valor es correcto y concatenar las nuevas columnas al DataFrame
for col in columnas_limpieza:
   
    split_cols = df_EDA[col].str.split(', ', expand=True)
    split_cols.columns = [f'{col}_1', f'{col}_2']  # Renombrar las nuevas columnas
    new_columns = pd.concat([new_columns, split_cols], axis=1)
#Concatenar new_columns a df_EDA
df_EDA = pd.concat([df_EDA, new_columns], axis=1)
```

```python
#Validacion del dataframe
df_EDA.head(5)
```

```python
#Validacion de valores
df_EDA[['clinch_att','clinch_att_1','clinch_att_2']].head(5)
```

```python
df_EDA[['total_str_att','total_str_att_1','total_str_att_2']].tail(5)
```

```python
#Eliminar columnas con valor original para evitar duplicar informacion
df_EDA.drop(columns=columnas_limpieza, inplace=True)
```
